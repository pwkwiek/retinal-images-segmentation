```python
import os
import cv2 

main_directory = ('C:/Users/User/Desktop/Isengard/data2')
sets = ['img','mk','seg']
types = ['test','train','val']
color = ['rgb_b','rgb_g','rgb_r','gray']

#create directories
for s in sets:
    for t in types:
        for c in color:
            try:
                os.makedirs(main_directory + '/chan/'+ str(c)+'/'+str(s)+'/'+str(t)+'/')
            except:
                print('exists')
```

```python
color = ['rgb_b','rgb_g','rgb_r']

for s in sets:
    for t in types:
        # assign directory
        directory = main_directory+'/'+str(s)+'/'+str(t)

        iter=-1
        # iterate over files in directory
        for filename in os.listdir(directory):
            f = os.path.join(directory, filename)
            # checking if it is a file
            if os.path.isfile(f):

                #read image
                image=cv2.imread(f)
                image_g=cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

                #split the image into its three channels
                channels = cv2.split(image)

                i=-1
                for c in color:
                    # Using cv2.imwrite() method 
                    # Saving the image 
                    file = main_directory + '/chan/'+str(c)+'/'+str(s)+'/'+str(t)+'/'+filename
                    cv2.imwrite(file, channels[i])
                    i=i+1
                
                file = main_directory + '/chan/gray/'+str(s)+'/'+str(t)+'/'+filename
                cv2.imwrite(file, image_g)
```
